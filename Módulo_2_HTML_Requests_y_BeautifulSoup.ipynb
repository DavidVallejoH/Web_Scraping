#PARSEANDO HTML CON **BEAUTIFULSOUP**

```
from bs4 import BeautifulSoup
sopa = BeautifulSoup(paginaweb.text, 'html.parser')
#prettify lo que hace es presentarlo de mejor manera a lo que hace .text
#print(sopa.prettify())
```

```
#resultado_sopa_encabezado = BeautifulSoup.find(ETIQUETA)
#find devuelve el primer ul de toda la pagina
a = sopa.find('ul')
a
```

```
#agrego atributos para especificar cual deseo
#agregando un diccionario especeficiando lo que busco y deseo obtener
#Find 'el' en el diccionario donde este class y hot sections
#find tambien todas las etiquetas li 
secciones = sopa.find('ul', attrs={'class':'hot-sections'}).find_all('li')
secciones
```

```
seccion = secciones[0]
seccion
```
